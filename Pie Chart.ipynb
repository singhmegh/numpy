```python
import matplotlib.pyplot as plt

brands = ["Oneplus","ios","Apple","Nokia","Redmi"]
x=[20,87,54,27,64]
c = ["red","blue","green","orange","Yellow"]
ex=[0,0,0,0,0.1]
plt.pie(x,labels = brands ,colors=c,explode =ex,shadow=True,autopct="% f",startangle=180)
plt.show()
```
