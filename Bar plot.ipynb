```python
import matplotlib.pyplot as plt

y=[98,76,46,73,29]
x=["Part1","Part2","Part3","Part4","Part5"]
colors=["red","yellow","blue","pink","orange"]
plt.bar(x,y,color=colors,edgecolor = "black")
plt.xlabel("Parts of harry potter")
plt.ylabel("Popularity",fontsize=17)
plt.title("popularity of differents parts of harry potter")
plt.show()
```
