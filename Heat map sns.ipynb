```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data=sns.load_dataset("tips")
print(data)
grouped_by=data.groupby("day").agg({"tip":"median"})
sns.heatmap(grouped_by)
plt.show()
```
